## Problem 1

The 2-stage Milne's method is

$$u_{n+1}=u_{n-1}+h\left(\frac{1}{3}f_{n+1}+\frac{4}{3}f_n+\frac{1}{3}f_{n-1}\right).$$

Prove with analytic computation that 

1. It is a 4th order method.

2. The region of absolute stability of is only along the imaginary axis.

**Solution**
1. It is a 4th order method

We have by analytic computation, the LTE:
\begin{equation}
    \tau_{n+1} = u(t_{n+1}) - u(t_{n-1}) - h( \frac{1}{3}f(u(t_{n+1})) + \frac{4}{3}f(u(t_n)) + \frac{1}{3}f(u(t_{n-1})))
\end{equation}

for simplicity, let $ u(t_{n-1}) = u $ and therefore $ f(u(t_{n-1})) = u'(t_{n-1}) = u' $. Using Taylor expansion:
\begin{equation}
    \tau_{n+1} =
                u + 2hu' + \frac{1}{2}(2h)^2u^{(2)}
              + \frac{1}{3!}(2h)^3u^{(3)} 
              + \frac{1}{4!}(2h)^4u^{(4)}
              + O(h^5) - u
              - h(\frac{1}{3}u'(t_{n+1}) + \frac{4}{3} u'(t_n)) + \frac{1}{3} u'(t_{n-1}))
\end{equation}

We can respectively expand $ u'(t_{n+1}) $ and $ u'(t_n) $.
\begin{equation}
    u'(t_{n+1}) = u' + 2h u^{(2)} + \frac{1}{2} (2h)^2 u^{(3)} + \frac{1}{3!} (2h)^3 u^{(4)} 
                  + \frac{1}{4!} (2h)^4 u^{(5)} + O(h^5)
\end{equation}

\begin{equation}
    u'(t_n) = u' + hu^{(2)} + \frac{1}{2} h^2 u^{(3)} + \frac{1}{3!} h^3 u^{(4)}
              + \frac{1}{4!} h^4 u^{(5)} + O(h^5)
\end{equation}

Plug back, combine and collect terms:
\begin{equation}
    \tau_{n+1} = 2hu' + 2h^2u^{(2)} + \frac{4}{3}h^3u^{(3)} + \frac{2}{3}h^4u^{(4)} + O(h^5)
               - \frac{1}{3}hu' - \frac{2}{3}h^2u^{(2)} - \frac{2}{3}h^3u^{(3)} - \frac{4}{9}h^4u^{(4)}
               - O(h^5) - \frac{4}{3}hu' - \frac{4}{3}h^2u^{(2)} - \frac{2}{3}h^3u^{(3)} - \frac{2}{9}h^4u^{(4)}
               - O(h^5) - \frac{1}{3}hu' \\
               = O(h^5)
\end{equation}
Then we have the order of the Milne-2 method is order 4.

**Solution**
(2) The region of absolute stability of it is only the imaginary axis.
The RAS for a LMM scheme is the set of points $z$ and $z \in \mathbb{C}$, such that plugging in $z$ satisfies the root condition for $p(w) - z\sigma(w)$.

For the Milne 2-stage method:
\begin{equation}
    u_{n+1} - u_{n-1} = h(\frac{1}{3}f(u_{n+1}) + \frac{4}{3} f(u_n) + \frac{1}{3} f(u_{n-1}))
\end{equation}
Yiedling characteristic polynomials: $p(w) = w^2 - 1$ and $\sigma(w) = \frac{1}{3}w^2 + \frac{4}{3}w + \frac{1}{3}$.
Explicitly solving $p(w) - z\sigma(w) = 0$ leads to complex manipulations. We can apply boundary locus method by assuming that a root is of the form $w_j = e^{i\theta}$ as it is on the "boundary".

Therefore, our polynomial becomes:
\begin{equation}
    p(e^{i\theta}) - z \sigma(e^{i\theta}) = 0 \\
\end{equation}
Therefore $z=\frac{p(e^{i\theta})}{\sigma(e^{i\theta})}$ can be explicitly calculated:
\begin{equation}
    z = \frac{e^{2i\theta}-1}{\frac{1}{3}e^{2i\theta} + \frac{4}{3}e^{i\theta} + \frac{1}{3}} \\
      = \frac{e^{i\theta}-e^{-i\theta}}{\frac{1}{3}e^{i\theta} + \frac{4}{3} + \frac{1}{3}e^{-i\theta}}
\end{equation}

Using the property of Euler's formula: $e^{i\theta} = \cos(\theta) + \sin(\theta)$ and $e^{-i\theta} = \cos(\theta) - \sin(\theta)$. We can simplify:
\begin{equation}
    z = \frac{3i\sin{\theta}}{\cos{\theta} + 2}
\end{equation}

Then our $RAS := \{
    z = \frac{e^{2i\theta}-1}{\frac{1}{3}e^{2i\theta} + \frac{4}{3}e^{i\theta} + \frac{1}{3}} = \frac{3i\sin{\theta}}{\cos{\theta} + 2} \big{|} \space \space 0 \le \theta \le 2\pi
\}$.

For any value: $0 \le \theta \le 2\pi$, the $i$ component in our numerator will suggest that the result is purely imaginary. Namely, since the numerator is always imaginary, and the denominator is always real, we have our $z \in \mathbb{C}$ for all $\theta$, namely the RAS is on the imaginary axis only.

## Problem 3

For the inhomogeneous linear equation.

$$u'(t)=\lambda u(t) + a,\quad a,u(t)\in\mathbb{R}.$$

Consider the Runge-Kutta method with Butcher tableau $A\in\mathbb{R}^{r\times r},b\in\mathbb{R}^{r},Re\lambda<0$

a) Prove that the numerical solution can be written as

$$u_n = R(\lambda h)^n u_0 + \left[\frac{R(\lambda h)^n-1}{R(\lambda h)-1}\right] h b^T(I-zA)^{-1}e a.$$

Here $e=(1,\ldots,1)^T\in\mathbb{R}^{r}$,

$$R(z)=1+zb^T(I-zA)^{-1}e.$$

The general Runge-Kutta method can be expressed as:

\begin{equation}
    k_i = f(u_n + \sum_j a_{ij} k_j) \\
    u_{n+1} = u_n + h\sum_i b_ik_i
\end{equation}

Or, in vector form:

\begin{equation}
    k = f(u_n e + hAk) \\
    u_{n+1} = u_n + hb^{T}k
\end{equation}

Begin our calculation:
\begin{equation}
    k=\lambda(u_n e+h A k) + ae \\
     =\lambda u_ne + \lambda hAk + ae
\end{equation}

Collecting terms, we have:
\begin{equation}
    (I-h\lambda A)k = \lambda eu_n + ea \\
    k = \lambda(I-h \lambda A)^{-1}eu_n + (I- h \lambda A)^{-1}ea
\end{equation}

Substituting $k$ into $u_{n+1} = u_n + hb^{T}k$, we obtain:

\begin{equation}
    u_{n+1} = u_n + hb^T(\lambda (I-h \lambda A)^{-1}eu_n + (I-h\lambda A)^{-1}ea) \\
    = u_n + h\lambda b^T(I-h\lambda A)^{-1}eu_n + hb^T(I-h\lambda A)^{-1}ea
\end{equation}

We let $z = h\lambda$, and substitute:
\begin{equation}
    u_{n+1} = u_n + zb^T(I-zA)^{-1}eu_n + hb^T(I-z A)^{-1}ea \\
    = (1 + zb^T(I-zA)^{-1}e)u_n + hb^T(I-z A)^{-1}ea \\
    = R(z)u_n + hb^T(I-z A)^{-1}ea
\end{equation}

Using this recursive relation, we can have:
\begin{equation}
    u_n = R(z)u_{n-1} + hb^T(I-z A)^{-1}ea \\
    = R(z) (R(z)u_{n-2} + hb^T(I-z A)^{-1}ea) + hb^T(I-z A)^{-1}ea = R(z)^2 u_{n-2} + (1+R(z))(hb^T(I-z A)^{-1}ea) \\
    = \cdots \\
    = R(z)^n u_0 + (1+R(z)+R(z)^2+\cdots + R(z)^{n-1})(hb^T(I-z A)^{-1}ea)
\end{equation}

Having the property of the geometric series, we can write $u_n$ compactly as:
\begin{equation}
    u_n = R(z)^n u_0 + \left[\frac{R(z)^n-1}{R(z)-1}\right] h b^T(I-zA)^{-1}e a.
\end{equation} which is proven by our calculation above.

b) Assume $|R(z)|<1$, prove that 

$$\lim_{n\to\infty} u_n=-\frac{a}{\lambda}.$$

Prove that the true solution of the ODE has the same limit, i.e.

$$\lim_{t\to+\infty} u(t)=-\frac{a}{\lambda}.$$

**(a)**
By algebraic manipulation:
\begin{equation}
    u_n = R(z)^n u_0 + \left[ \frac{1}{R(z) - 1}R(z)^n - \frac{1}{R(z) - 1}\right]hb^T(I-zA)^{-1}ea \\
    = R(z)^n u_0 + \left( \frac{hb^T(I-zA)^{-1}ea}{R(z)-1} \right)R(z)^n - \frac{hb^T(I-zA)^{-1}ea}{R(z)-1} \\
    = R(z)^n \left( u_0 + \frac{hb^T(I-zA)^{-1}ea}{R(z)-1}\right) - \frac{hb^T(I-zA)^{-1}ea}{R(z)-1}
\end{equation} notice except for $R(z)^n$ that is explicitly dependent on $n$, all other terms are constants.

Taking the limit:
\begin{equation}
    \lim_{n\to\infty} u_n = C_1\lim_{n\to\infty}R(z)^n - C_2
\end{equation} where $C_1, C_2$ are, respectively, $ u_0 + \frac{hb^T(I-zA)^{-1}ea}{R(z)-1} $ and $\frac{hb^T(I-zA)^{-1}ea}{R(z)-1}$.

We obtain, because of the convergence of $R(z)^n$ since $|R(z)|<1$, we must have $\lim_{n\to\infty}R(z)^n = 0$

\begin{equation}
    \lim_{n\to\infty} u_n = C_1 \bullet 0 - C_2 = -C_2 \\
    = - \frac{hb^T(I-zA)^{-1}ea}{R(z)-1}
\end{equation}

but $$R(z)=1+zb^T(I-zA)^{-1}e.$$

Plugging in, we have:
\begin{equation}
    \lim_{n\to\infty} u_n = - \frac{hb^T(I-zA)^{-1}ea}{1+zb^T(I-zA)^{-1}e -1} \\
    = - \frac{hb^T(I-zA)^{-1}ea}{zb^T(I-zA)^{-1}e} \\
    = - \frac{hb^T(I-zA)^{-1}ea}{h\lambda b^T(I-zA)^{-1}e} \\
\end{equation} This suggests:
\begin{equation}
    \lim_{n\to\infty} u_n = -\frac{a}{\lambda}
\end{equation} after cancelling terms.



**(b)**
Using Duhammel's principle, we can integrate our exact solution explicitly:
\begin{equation}
    u(t) = e^{A(t)}u_0 + e^{A(t)} \int_{0}^{t} e^{-A(s)}b(s) ds
\end{equation}
where from our original ODE: $u' = \lambda u + a$ we have $a(t) = \lambda , b(t) = a$.

Then we have the formula:
\begin{equation}
    u(t) = e^{A(t)}u_0 + e^{A(t)}\int_0^{t} e^{-A(s)}b(s)ds \\
    = e^{\lambda t}u_0 + e^{\lambda t} a\int_0^t e^{-\lambda t} ds \\
    = e^{\lambda t}u_0 + ae^{\lambda t} (-\frac{1}{\lambda}e^{-\lambda s}\Bigg{|}_{0}^{t}) \\
    = e^{\lambda t}u_0 + ae^{\lambda t}(\frac{1}{\lambda} - \frac{1}{\lambda}e^{-\lambda t}) \\
    = e^{\lambda t}(u_0 + \frac{a}{\lambda}) - \frac{a}{\lambda}
\end{equation}

Obviously since we have $Re\lambda<0$, we must have $\lim_{t\to +\infty} e^{\lambda t} = 0$, namely:
\begin{equation}
    \lim_{t\to\infty} u(t) = (u_0 + \frac{a}{\lambda})(\lim_{t\to\infty} e^{\lambda t}) - \frac{a}{\lambda} \\
    = 0 - \frac{a}{\lambda} = -\frac{a}{\lambda}
\end{equation} proving our statement.

(1) Obtain the solution $u(x,T)$ at $T=1.0$ using an adaptive time stepping strategy (with the "Milne device").  The initial time step is $h=0.001$. Plot out the history of the size step along the history of the simulation.

Due to space discretization, we can view the partial derivative with respect to time as a single function with a fixed x at each grid point, therefore we have the system of equations in $t$:
\begin{equation}
\dot{u}_0 (t) = u_t(0, t) = \frac{1}{\Delta x^2} (u_1 - 2u_0 + u_{-1}) \\
\dot{u}_1 (t) = u_t(\Delta x, t) = \frac{1}{\Delta x^2} (u_2 - 2u_1 + u_0) \\
\dot{u}_2 (t) = u_t(2\Delta x, t) = \frac{1}{\Delta x^2} (u_3 - 2u_2 + u_1) \\ 
\vdots \\
\dot{u}_N (t) = u_t(N\Delta x, t) = \frac{1}{\Delta x^2} (u_{N+1} - 2u_N + u_{N-1})
\end{equation}
Because of the periodicity, we can replace $u_{-1}, u_{N+1}$ by $u_N, u_0$. Compactly, this can be represented by:
\begin{equation}
\mathbf{\dot{u}_t} = \frac{1}{\Delta x^2} \mathbf{A} \mathbf{u}
\end{equation}, where $\mathbf{A}$ is:

\begin{pmatrix}
    -2 & 1 & 0 & \dots & 0 & 1 \\
    1 & -2 & 1 & 0 & \dots & 0 \\
      & \ddots & \ddots & \ddots & \vdots & \vdots \\
      & & \ddots & \ddots & \ddots \\
      0 & \dots & 0 & 1 & -2 & 1 \\
      1 & 0 & \dots & 0 & 1 & -2
\end{pmatrix}, implementing the periodic boundary conditions:
\begin{equation}
\quad u(0,t)=u(2\pi,t), u_x(0,t)=u_x(2\pi,t)
\end{equation}

Therefore our $f$ on the right hand side is simply the matrix multiplication at each time step ($f$ is not explicitly time-dependent):
\begin{equation}
    \mathbf{\dot{u}} = \mathbf{f(u)} = \frac{1}{\Delta{x}^2}A\mathbf{u}
\end{equation}

The two Runge-Kutta methods can be described as the following:
\begin{equation}
    k_1 = f(t_n,u_n) = f(u_n)\\
    k_2 = f(t_n + \frac{2}{3}h, u_n + h\frac{2}{3}k_1) = f(u_n + h\frac{2}{3}k_1)\\
    k_3 = f(t_n + \frac{2}{3}h, u_n + h\frac{2}{3}k_2) = f(u_n + h\frac{2}{3}k_2)\\
\end{equation}

With the propagator of order $p=2$:
\begin{equation}
    u_{n+1} = u_n + h(\frac{1}{4}k_1 + \frac{3}{4}k_2)
\end{equation}

And the error controller of order $\tilde{p}=3$:
\begin{equation}
    \tilde{u}_{n+1} = u_n + h(\frac{1}{4}k_1 + \frac{3}{8}k_2 + \frac{3}{8}k_3)
\end{equation}

We have error control per time step $\kappa = ||\tilde{u}_{n+1} - u_{n+1}
||$:
\begin{equation}
    \kappa = \frac{3}{8} ||f(t_n + \frac{2}{3}h, k_3) - f(t_n + \frac{2}{3}h, k_2) ||
\end{equation}

```julia
using LinearAlgebra
using SparseArrays
# Set up the problem
function initial(X)
    #= This is a helper function that takes in X a vector of discretized spatial points, and 
    returns a vector representing u_0 our initial value at t=0 =#
    return exp.(((-5) * ((X .- pi).^2)))
end

#= Set up variables for the methods =#
h0 = 0.001; # initial time step
T = 1.0;
dx = (2 * pi) / 100;
discretizedSpaceArray = collect(0:dx:2*pi);
u0 = initial(discretizedSpaceArray); len_u = length(u0);
#####################
# Create sparse matrix representing the central finite difference rule
# Create entries in A
A = spzeros(len_u, len_u);
A[1, 1] = -2; A[1, 2] = 1; A[1, end] = 1;
A[end, 1] = 1; A[end, end - 1] = 1; A[end, end] = -2;
for row in 2:(len_u-1)
    A[row, row-1] = 1;
    A[row, row] = -2;
    A[row, row+1] = 1;
end
# A loaded successfully

function func(u, dx)
    #= A helper function representing the right hand side of u' = F(u) =#
   return (1/(dx)^2) .* (A * u)
end
```

```julia
using LinearAlgebra
using SparseArrays

# set up tolerance
tol = Float64(10^(-2))

function RK_propagator_controller_onestep(h, u0)
    #= Due to the high similarity of the two RK methods, we can compute the propagator and
    controller simultaneously during each call, can also directly calculate kappa =#
    k1 = func(u0, dx)
    k2 = func(u0 + (2/3)*h*k1, dx)
    k3 = func(u0 + (2/3)*h*k2, dx)
    propagator = u0 + h*((1/4) * k1 + (3/4) * k2)
    monitor = u0 + h*((1/4) * k1 + (3/8) * k2 + (3/8) * k3)
    kappa = (3/8) * norm(func(k3, dx) - func(k2, dx))
    return propagator, monitor, kappa

end
##########################################

function Embeded_RK(h, T, u0, tol)
    H = Float64[h]; U = u0; t = 0; Times = Float64[t]
    h1 = h
    w1 = U[:,end]
    while (t <= T) 
        y_new, y_monitor, k = RK_propagator_controller_onestep(h1, w1)
        if k > h*tol
            h1 = (1/2) * h1
            continue # no need to remesh
        else
            if t + h1 < T
                if k > (1/10) * h*tol
                    t = t + h1; U = [U y_new]
                    append!(H, h1); append!(Times, t)
                    w1 = U[:, end];
            
                    continue # solution survives, propagate
                else
                    t = t + h1; U = [U y_new]
                    append!(H, h1); append!(Times, t)
                    h1 = 2 * h1
                    w1 = U[:, end];
                    
                    continue # doubled stepsize, no need to remesh
                    end # end if: check double h
            else
                break
                end # end if: check t > T
            loop += 1
            end # end if check halve h
        end # end while loop
    return U[:, end], H, Times
    end # end function Embeded_RK

sol, stepSizes, timeTrack = Embeded_RK(h0, T, u0, tol);
```

```julia

# Plotting step size change
using PyPlot
figure(1,figsize=(5,5))
plot(timeTrack, stepSizes,"r-",label="Step Size Change")
xlabel("t")
ylabel("Step size h")
legend()
title(L"ERK, \delta = 0.01,t,h")
```

```julia
###########################################
# Plotting solution distribution at x=2pi
X = [0:2*pi/100:2*pi;]
figure(1,figsize=(10,5))
plot(X, sol, "b-",label="Solution at Discretized Grid")
xlabel("Discretized x")
ylabel("u(x, T)")
legend()
grid("on")
title(L"ERK,t=T")
```
